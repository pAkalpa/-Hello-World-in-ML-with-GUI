```python
import pandas as pd

dataset = pd.read_csv('iris_data.csv').values
print(dataset,dataset.shape)
```

```python
data = dataset[:,:4]
target = dataset[:,4]
print(data)
```

```python
print(target)
```

```python
from sklearn.model_selection import train_test_split as tts
train_data,test_data,train_target,test_target=tts(data,target,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(train_data,train_target)
```

```python
predicted_target = model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test_target,predicted_target)
print('Accuracy:'+str(acc * 100)+'%')
```

```python
testmodel_data = [[5.2,1.2,4.2,0.5]]

result = model.predict(testmodel_data)
print(result)
```

```python
import joblib

joblib.dump(model,'IRIS_KNN.sav')
```
